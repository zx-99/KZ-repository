# Tutorial 5:   Education by Country

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
```

## Loading the data with pandas


```python
df=pd.read_csv('education-by-country.csv', thousands=',')
df.head()
```

## Using Pie Chart

```python
size_of_groups=df['t_education_percent']

plt.pie(size_of_groups)

plt.show()
```

## Adding More Information  


```python
names=df['country']

size_of_groups=df['t_education_percent']

plt.pie(size_of_groups)

plt.pie(size_of_groups, labels=names,labeldistance=1.15, autopct='%1.0f%%')

plt.show()
```
